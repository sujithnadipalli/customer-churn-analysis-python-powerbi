```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")

print("Libraries Imported Successfully")
```

```python
df = pd.read_excel("Customer_churn_Analysis_dataset.xlsx")

df.head()
```

```python
print("Rows and Columns:", df.shape)

df.info()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)

print("Duplicates Removed")
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Exited',
    data=df
)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Geogrphy',
    hue='Exited',
    data=df
)

plt.title("Churn by Geography")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Gender',
    hue='Exited',
    data=df
)

plt.title("Churn by Gender")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Age'],
    bins=30,
    kde=True
)

plt.title("Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Balance'],
    bins=30,
    kde=True
)

plt.title("Balance Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='IsActiveMember',
    hue='Exited',
    data=df
)

plt.title("Churn by Active Status")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='NumOfProducts',
    hue='Exited',
    data=df
)

plt.title("Churn by Number of Products")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['CreditScore'],
    bins=25,
    kde=True
)

plt.title("Credit Score Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x='Exited',
    y='Age',
    data=df
)

plt.title("Age vs Churn")

plt.show()
```
